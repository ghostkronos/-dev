```python
def en_buyuk(liste):
    return max(liste)

print(en_buyuk([1, 2, 3, 4, 5]))
print(en_buyuk([-1, -5, -3]))
```

```python
def ters_cevir(s):
    return s[::-1]

print(ters_cevir("hello"))
print(ters_cevir("Python"))
```

```python
def faktoriyel(n):
    sonuc = 1
    for i in range(1, n+1):
        sonuc *= i
    return sonuc

print(faktoriyel(5))
print(faktoriyel(0))
```

```python
def ciftleri_bul(liste):
    return [x for x in liste if x % 2 == 0]

print(ciftleri_bul([1, 2, 3, 4, 5, 6]))
print(ciftleri_bul([11, 23, 44]))
```

```python
def birlestir_tekil(l1, l2):
    return list(set(l1 + l2))

print(birlestir_tekil([1,2,3], [3,4,5]))
print(birlestir_tekil([], [1,1,2]))
```

```python
def palindrom_mu(s):
    s = s.replace(" ", "").lower()
    return s == s[::-1]

print(palindrom_mu("A man a plan a canal Panama"))
print(palindrom_mu("hello"))
```

```python
def ebob(a, b):
    while b != 0:
        a, b = b, a % b
    return a

print(ebob(54, 24))
print(ebob(7, 13))
```

```python
def anagram_mi(s1, s2):
    s1 = s1.replace(" ", "").lower()
    s2 = s2.replace(" ", "").lower()
    return sorted(s1) == sorted(s2)

print(anagram_mi("listen", "silent"))
print(anagram_mi("hello", "world"))
```

```python
def eksik_sayi(liste):
    n = len(liste) + 1
    toplam = n * (n + 1) // 2
    return toplam - sum(liste)

print(eksik_sayi([1,2,4,5]))
```

```python
def dondur(liste, k):
    k = k % len(liste)
    return liste[-k:] + liste[:-k]

print(dondur([1,2,3,4,5], 2))
print(dondur([-1,-100,3,99], 2))
```

```python
def median_sorted(a, b):
    import statistics
    return statistics.median(a+b)

print(median_sorted([1,3], [2]))
print(median_sorted([1,2], [3,4]))
```

```python
def longest_palindrome(s):
    sonuc = ""
    for i in range(len(s)):
        for j in range(i, len(s)):
            parca = s[i:j+1]
            if parca == parca[::-1] and len(parca) > len(sonuc):
                sonuc = parca
    return sonuc

print(longest_palindrome("babad"))
print(longest_palindrome("cbbd"))
```

```python
from collections import deque

def word_ladder(begin, end, wordList):
    wordSet = set(wordList)
    if end not in wordSet:
        return 0
    queue = deque([(begin, 1)])
    while queue:
        kelime, seviye = queue.popleft()
        if kelime == end:
            return seviye
        for i in range(len(kelime)):
            for c in "abcdefghijklmnopqrstuvwxyz":
                yeni = kelime[:i] + c + kelime[i+1:]
                if yeni in wordSet:
                    wordSet.remove(yeni)
                    queue.append((yeni, seviye+1))
    return 0

print(word_ladder("hit","cog",["hot","dot","dog","lot","log","cog"]))
print(word_ladder("hit","cog",["hot","dot","dog","lot","log"]))
```

```python
def max_subarray(liste):
    max_ = cur = liste[0]
    for x in liste[1:]:
        cur = max(x, cur + x)
        max_ = max(max_, cur)
    return max_

print(max_subarray([-2,1,-3,4,-1,2,1,-5,4]))
print(max_subarray([5,4,-1,7,8]))
```

```python
def merge(intervals):
    intervals.sort()
    sonuc = [intervals[0]]
    for bas, bit in intervals[1:]:
        if bas <= sonuc[-1][1]:
            sonuc[-1][1] = max(sonuc[-1][1], bit)
        else:
            sonuc.append([bas, bit])
    return sonuc

print(merge([[1,3],[2,6],[8,10],[15,18]]))
print(merge([[1,4],[4,5]]))
```
